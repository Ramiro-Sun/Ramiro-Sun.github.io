```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the ranges for u1 and u2 and a small increment for sigma
u1 = np.linspace(-2, 2, 400)
u2 = np.linspace(-2, 2, 400)
sigma = np.linspace(1, 3, 10)

# Create the meshgrid for u1 and u2
U1, U2 = np.meshgrid(u1, u2)
Radius = np.sqrt(U1**2 + U2**2)

# Condition for the circle region 1 ≤ sqrt(u1^2 + u2^2) ≤ 2
mask = (Radius >= 1) & (Radius <= 2)

# For every valid (u1, u2), plot the circle with all sigma values
for s in sigma:
    ax.scatter(U1[mask], U2[mask], s, alpha=0.1)

ax.set_xlabel('u1')
ax.set_ylabel('u2')
ax.set_zlabel('sigma')
plt.title('3D plot of the set U1')
plt.show()
```

```python
u1 = np.linspace(-2, 2, 400)
u2 = np.linspace(-2, 2, 400)
sigma = np.linspace(1, 3, 10)  # Start sigma at 1 and go higher

# Create the meshgrid for u1, u2, and sigma
U1, U2 = np.meshgrid(u1, u2)
Radius = np.sqrt(U1**2 + U2**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# For each sigma, calculate the maximum radius allowed which is min(2, sigma)
for s in sigma:
    max_radius = min(2, s)
    mask = (Radius <= max_radius)
    ax.scatter(U1[mask], U2[mask], s, alpha=0.1)

ax.set_xlabel('u1')
ax.set_ylabel('u2')
ax.set_zlabel('sigma')
plt.title('3D plot of the set U2')
plt.show()
```
